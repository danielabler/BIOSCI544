```
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
%matplotlib inline
```

```
def plot_randomWalk(S,tmax):
  fig, ax1 = plt.subplots(1, 1)
  plt.plot(S[0],'ro')
  for t in range(tmax):
    s=2*(np.random.rand()>0.5)-1
    S.append(s)
  if len(S)<10:
    plt.plot(np.cumsum(S),'ro-')
  else:
    plt.plot(np.cumsum(S),'r-')
  print(S)
  return S


```

```
def randomWalk(tmax):
  return np.cumsum(2*(np.random.rand(tmax)>0.5)-1)
def stat(rep,tmax):
  S=np.array([randomWalk(tmax) for r in range(rep)])
  D2=(S**2).mean(0)
  D=np.sqrt(D2)
  return D
D=stat(10000,1000)
plt.plot(np.arange(D.size),D,'.')
plt.plot(np.arange(D.size),np.sqrt(np.arange(D.size)));
```
